## üìå –®–ø–∞—Ä–≥–∞–ª–∫–∞ (–Ω–∞ —Å–µ–≥–æ–¥–Ω—è)

**–¶–∏–∫–ª—ã –ø–æ —Å—Ç—Ä–æ–∫–∞–º –∏ —Å–ø–∏—Å–∫–∞–º**
- `for x in seq:` ‚Äî –ø—Ä–æ–π—Ç–∏—Å—å –ø–æ **–∫–∞–∂–¥–æ–º—É** —ç–ª–µ–º–µ–Ω—Ç—É (—Å—Ç—Ä–æ–∫–∏, —Å–ø–∏—Å–∫–∏).
- –ü–æ –∏–Ω–¥–µ–∫—Å–∞–º: `for i in range(len(seq)):` ‚Äî –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–∞ –ø–æ–∑–∏—Ü–∏—è.
- `enumerate(seq, start=1)` ‚Äî –ø–∞—Ä—ã `(–Ω–æ–º–µ—Ä, —ç–ª–µ–º–µ–Ω—Ç)` –¥–ª—è —á–µ–ª–æ–≤–µ–∫–æ‚Äë—á–∏—Ç–∞–µ–º–æ–π –Ω—É–º–µ—Ä–∞—Ü–∏–∏.

**`range()` ‚Äî –≥–µ–Ω–µ—Ä–∞—Ü–∏—è –∏–Ω–¥–µ–∫—Å–æ–≤**
- `range(n)` ‚Üí `0..n-1`, `range(a, b)`, `range(a, b, step)` (—à–∞–≥ –º–æ–∂–µ—Ç –±—ã—Ç—å –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–º).
- ¬´–ö–∞–∂–¥—ã–π N‚Äë–π¬ª: `for i in range(0, len(tokens), N): ...`

**–£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ü–∏–∫–ª–æ–º**
- `continue` ‚Äî –ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å —Ç–µ–∫—É—â—É—é –∏—Ç–µ—Ä–∞—Ü–∏—é.  
- `break` ‚Äî –≤—ã–π—Ç–∏ –∏–∑ —Ü–∏–∫–ª–∞ –ø–æ–ª–Ω–æ—Å—Ç—å—é.  
- `pass` ‚Äî –∑–∞–≥–ª—É—à–∫–∞ (–Ω–∏—á–µ–≥–æ –Ω–µ –¥–µ–ª–∞–µ—Ç).  
- `for ... else` ‚Äî `else` –≤—ã–ø–æ–ª–Ω–∏—Ç—Å—è **–µ—Å–ª–∏ `break` –Ω–µ —Å—Ä–∞–±–æ—Ç–∞–ª**.

**`while` ‚Äî ¬´–ø–æ–∫–∞ —É—Å–ª–æ–≤–∏–µ –∏—Å—Ç–∏–Ω–Ω–æ¬ª**
- –ë–µ—Ä–∏—Ç–µ, –∫–æ–≥–¥–∞ **–Ω–µ –∑–Ω–∞–µ—Ç–µ –∑–∞—Ä–∞–Ω–µ–µ**, —Å–∫–æ–ª—å–∫–æ –∏—Ç–µ—Ä–∞—Ü–∏–π, –∏–ª–∏ —à–∞–≥ **–∑–∞–≤–∏—Å–∏—Ç –æ—Ç –¥–∞–Ω–Ω—ã—Ö**.  
- –ë–µ–∑–æ–ø–∞—Å–Ω–æ—Å—Ç—å: –≤–Ω—É—Ç—Ä–∏ —Ü–∏–∫–ª–∞ **–º–µ–Ω—è–π—Ç–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ** (—Å—á—ë—Ç—á–∏–∫/–∏–∑–º–µ–Ω–µ–Ω–∏–µ —Å—Ç—Ä–æ–∫–∏/–ø–æ–ø–æ–ª–Ω–µ–Ω–∏–µ —Å–ø–∏—Å–∫–∞).  
- –û–≥—Ä–∞–Ω–∏—á–∏—Ç–µ–ª—å: —Å—á—ë—Ç—á–∏–∫ —à–∞–≥–æ–≤ –¥–ª—è –∑–∞—â–∏—Ç—ã –æ—Ç ¬´–≤–µ—á–Ω—ã—Ö¬ª —Ü–∏–∫–ª–æ–≤.

**–°–ª–æ–∂–Ω—ã–µ —É—Å–ª–æ–≤–∏—è**
- –ü—Ä–∏–æ—Ä–∏—Ç–µ—Ç: `not` > `and` > `or` ‚Üí —Å—Ç–∞–≤—å—Ç–µ **—Å–∫–æ–±–∫–∏**.  
- –î–µ –ú–æ—Ä–≥–∞–Ω: `not (A and B)` ‚áî `(not A) or (not B)`; `not (A or B)` ‚áî `(not A) and (not B)`.

**–ê–Ω—Ç–∏‚Äë–∫–∞–ø–∫–∞–Ω—ã**
- –ù–µ –º–æ–¥–∏—Ñ–∏—Ü–∏—Ä—É–π—Ç–µ —Å–ø–∏—Å–æ–∫, –ø–æ –∫–æ—Ç–æ—Ä–æ–º—É –∏—Ç–µ—Ä–∏—Ä—É–µ—Ç–µ—Å—å (–¥–µ–ª–∞–π—Ç–µ –∫–æ–ø–∏—é –∏–ª–∏ –Ω–æ–≤—ã–π —Å–ø–∏—Å–æ–∫).
- `range(a, b)` **–Ω–µ –≤–∫–ª—é—á–∞–µ—Ç** `b`.
- –ò–∑–±–µ–≥–∞–π—Ç–µ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ –≤ `while`: –∏–∑–º–µ–Ω—è–π—Ç–µ —Å–æ—Å—Ç–æ—è–Ω–∏–µ –∏/–∏–ª–∏ –¥–æ–±–∞–≤–ª—è–π—Ç–µ –æ–≥—Ä–∞–Ω–∏—á–∏—Ç–µ–ª—å.

```python
# –ú—É–∂—Å–∫–∏–µ vs –∂–µ–Ω—Å–∫–∏–µ –∏–º–µ–Ω–∞ (–ø—Ä–æ—Å—Ç–µ–π—à–∏–π –∞–Ω–∞–ª–∏–∑)
female_endings = ['–∞', '—è']
likely_female = [name for name in characters_names
                if any(name.split()[0].endswith(ending) for ending in female_endings)]
print("‚ôÄÔ∏è  –í–µ—Ä–æ—è—Ç–Ω–æ –∂–µ–Ω—Å–∫–∏–µ –∏–º–µ–Ω–∞:", likely_female)
```

```python
female_endings = ['–∞', '—è']

likely_female = []

for name in characters_names:
  parts = name.split()
  has_female_ending = False
  for part in parts:
    if part.endswith(tuple(female_endings)):
      has_female_ending = True
      break
  if has_female_ending:
    likely_female.append(name)

print("‚ôÄÔ∏è  –í–µ—Ä–æ—è—Ç–Ω–æ –∂–µ–Ω—Å–∫–∏–µ –∏–º–µ–Ω–∞:", likely_female)
```

```python
for name in likely_female:
    print(f"{name} (–¥–ª–∏–Ω–∞ {len(name)}, —Å–ª–æ–≤ {name.count(' ') + 1})")
```

### –ü—Ä–∏–æ—Ä–∏—Ç–µ—Ç –æ–ø–µ—Ä–∞—Ç–æ—Ä–æ–≤
`not` –≤—ã—à–µ `and` –≤—ã—à–µ `or`. **–í—Å–µ–≥–¥–∞ —Å—Ç–∞–≤—å—Ç–µ —Å–∫–æ–±–∫–∏**, –µ—Å–ª–∏ —Å–æ–º–Ω–µ–≤–∞–µ—Ç–µ—Å—å.

**De Morgan:**
- `not (A and B)` ‚áî `(not A) or (not B)`
- `not (A or B)` ‚áî `(not A) and (not B)`


```python
text = "–ù–æ—á—å, —É–ª–∏—Ü–∞, —Ñ–æ–Ω–∞—Ä—å, –∞–ø—Ç–µ–∫–∞...\n–ë–µ—Å—Å–º—ã—Å–ª–µ–Ω–Ω—ã–π –∏ —Ç—É—Å–∫–ª—ã–π —Å–≤–µ—Ç."
is_night = '–Ω–æ—á' in text.lower()
is_light = "—Å–≤–µ—Ç" in text.lower()

if is_night or is_light:
    print("True")
```

```python
if is_night and is_light:
    print('–°–æ–¥–µ—Ä–∂–∏—Ç –Ω–æ—á—å –∏ —Å–≤–µ—Ç')
else:
    print('–ù–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –∏ –Ω–æ—á—å, –∏ —Å–≤–µ—Ç –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ')
```

```python
if ('–Ω–æ—á' in text.lower() and "–¥–æ–∂–¥—å" in text.lower()):
    print('–°–æ–¥–µ—Ä–∂–∏—Ç –Ω–æ—á—å –∏ –¥–æ–∂–¥—å')
elif '–Ω–æ—á' in text.lower():
    print('–°–æ–¥–µ—Ä–∂–∏—Ç –Ω–æ—á—å, –Ω–æ –Ω–µ –¥–æ–∂–¥—å')
elif "–¥–æ–∂–¥—å" in text.lower():
    print('–°–æ–¥–µ—Ä–∂–∏—Ç –¥–æ–∂–¥—å, –Ω–æ –Ω–µ –Ω–æ—á—å')
else:
    print('–ù–µ —Å–æ–¥–µ—Ä–∂–∏—Ç –∏ –Ω–æ—á—å, –∏ –¥–æ–∂–¥—å –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ')
```

## 4) –£–ø—Ä–∞–≤–ª–µ–Ω–∏–µ —Ü–∏–∫–ª–æ–º: `break`, `continue`, `pass`, `for ... else`

```python
# –ò—â–µ–º –ø–µ—Ä–≤–æ–µ —Å–ª–æ–≤–æ —Å 'amor'; –ø—Ä–æ–ø—É—Å–∫–∞–µ–º –∫–æ—Ä–æ—Ç–∫–∏–µ
found = None
for w in words:
    if len(w) < 4:
        continue
    if "amor" in w.lower():
        found = w
        print("–ù–∞—à–ª–∏:", w)
        break
else:
    print("–ù–µ –Ω–∞—à–ª–∏ 'amor'")
found
```

```python
# –£–º–Ω–∞—è –Ω–∞–≤–∏–≥–∞—Ü–∏—è –ø–æ —Ç–µ–∫—Å—Ç—É —Å —É–ø—Ä–∞–≤–ª—è—é—â–∏–º–∏ –∫–æ–º–∞–Ω–¥–∞–º–∏
text = "–ú–æ–π –¥—è–¥—è —Å–∞–º—ã—Ö —á–µ—Å—Ç–Ω—ã—Ö –ø—Ä–∞–≤–∏–ª, –∫–æ–≥–¥–∞ –Ω–µ –≤ —à—É—Ç–∫—É –∑–∞–Ω–µ–º–æ–≥"

print("=== BREAK: '–∑–∞–∫—Ä—ã—Ç—å –∫–Ω–∏–≥—É' (–ø–æ–ª–Ω–∞—è –æ—Å—Ç–∞–Ω–æ–≤–∫–∞) ===")
print("–ß–∏—Ç–∞–µ–º –¥–æ –ø–µ—Ä–≤–æ–π –∑–∞–ø—è—Ç–æ–π:")
result = ""
for char in text:
    if char == ',':
        print(f"‚û§ –ù–∞—à–ª–∏ –∑–∞–ø—è—Ç—É—é! –û—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º—Å—è.")
        break  # –≤—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞ –ø–æ–ª–Ω–æ—Å—Ç—å—é
    result += char
    print(f"–ß–∏—Ç–∞–µ–º: '{char}' (–Ω–∞–∫–æ–ø–ª–µ–Ω–æ: '{result}')")
print(f"–†–µ–∑—É–ª—å—Ç–∞—Ç –¥–æ –∑–∞–ø—è—Ç–æ–π: '{result}'")
```

```python
print("\n=== CONTINUE: '–ø—Ä–æ–ø—É—Å—Ç–∏—Ç—å –∞–±–∑–∞—Ü' (–ø—Ä–æ–ø—É—Å–∫ –∏—Ç–µ—Ä–∞—Ü–∏–∏) ===")
print("–°–æ–±–∏—Ä–∞–µ–º —Ç–µ–∫—Å—Ç –±–µ–∑ –ø—Ä–æ–±–µ–ª–æ–≤ –∏ –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è:")
clean_text = ""
for char in text:
    if char in ' ,.!?;:':
        print(f"‚û§ –ü—Ä–æ–ø—É—Å–∫–∞–µ–º: '{char}'")
        continue  # –ø—Ä–æ–ø—É—Å–∫–∞–µ–º –æ—Å—Ç–∞–ª—å–Ω—É—é —á–∞—Å—Ç—å –∏—Ç–µ—Ä–∞—Ü–∏–∏
    clean_text += char
    print(f"–î–æ–±–∞–≤–ª—è–µ–º: '{char}' (–Ω–∞–∫–æ–ø–ª–µ–Ω–æ: '{clean_text}')")
print(f"–û—á–∏—â–µ–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç: '{clean_text}'")
```

```python
print("\n=== PASS: '–∑–∞–º–µ—Ç–∫–∞ –¥–ª—è –±—É–¥—É—â–µ–≥–æ' (–≤—Ä–µ–º–µ–Ω–Ω–∞—è –∑–∞–≥–ª—É—à–∫–∞) ===")
words = text.split()
for word in words:
    if len(word) > 7:
        pass  # TODO: –¥–æ–±–∞–≤–∏—Ç—å –∞–Ω–∞–ª–∏–∑ –¥–ª–∏–Ω–Ω—ã—Ö —Å–ª–æ–≤ –ø–æ–∑–∂–µ
        print(f"üìù –î–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ '{word}' ‚Äî –∞–Ω–∞–ª–∏–∑ –æ—Ç–ª–æ–∂–µ–Ω")
    elif word.startswith('—Å'):
        print(f"‚úì –°–ª–æ–≤–æ –Ω–∞ '—Å': {word}")
    else:
        print(f"‚Üí –û–±—ã—á–Ω–æ–µ —Å–ª–æ–≤–æ: {word}")
```

```python
# –ë–µ–∑–æ–ø–∞—Å–Ω–∞—è ¬´–±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç—å¬ª —Å –æ–≥—Ä–∞–Ω–∏—á–∏—Ç–µ–ª–µ–º
count = 0
while True:
    count += 1
    if count > 5:
        break
count
```

```python
# While: —É—Å–ª–æ–≤–Ω–∞—è –∏—Ç–µ—Ä–∞—Ü–∏—è –¥–ª—è –≥—É–º–∞–Ω–∏—Ç–∞—Ä–∏–µ–≤
print("=== –ë–∞–∑–æ–≤—ã–π while: –ø—Ä–æ—Å—Ç–æ–π —Å—á—ë—Ç—á–∏–∫ ===")
n = 0
while n < 3:
    print(f"–®–∞–≥ {n}: –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º –¥–∞–Ω–Ω—ã–µ...")
    n += 1  # –ö–†–ò–¢–ò–ß–ù–û! –ó–∞–±—ã—Ç—å —ç—Ç–æ = –±–µ—Å–∫–æ–Ω–µ—á–Ω—ã–π —Ü–∏–∫–ª
print("‚úì –û–±—Ä–∞–±–æ—Ç–∫–∞ –∑–∞–≤–µ—Ä—à–µ–Ω–∞")
```

```python
pending = ["doc1", "doc2", "doc3"]
processed = []

while pending:
    doc = pending.pop(0)   # –≤–∑—è–ª–∏ –ø–µ—Ä–≤—ã–π –∏ —É–¥–∞–ª–∏–ª–∏
    doc = doc.upper()  # –æ–±—Ä–∞–±–æ—Ç–∫–∞ (–∑–∞–≥–ª—É—à–∫–∞)
    processed.append(doc)

processed  # ['DOC1', 'DOC2', 'DOC3']
```

```python
print("\n=== DH-–ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ: –ø–æ–∏—Å–∫ –≤ —Ç–µ–∫—Å—Ç–µ ===")
text = "–í –Ω–∞—á–∞–ª–µ –∏—é–ª—è, –≤ —á—Ä–µ–∑–≤—ã—á–∞–π–Ω–æ –∂–∞—Ä–∫–æ–µ –≤—Ä–µ–º—è..."
search_word = "–∂–∞—Ä–∫–æ–µ"
words = text.split()

# –ò—â–µ–º —Å–ª–æ–≤–æ, —Å–æ–¥–µ—Ä–∂–∞—â–µ–µ –∏—Å–∫–æ–º—É—é –ø–æ–¥—Å—Ç—Ä–æ–∫—É
index = 0
found = False
result_word = None

while index < len(words) and not found:
    current_word = words[index].strip(',.!?')  # —É–±–∏—Ä–∞–µ–º –ø—É–Ω–∫—Ç—É–∞—Ü–∏—é
    print(f"–ü—Ä–æ–≤–µ—Ä—è–µ–º —Å–ª–æ–≤–æ #{index}: '{current_word}'")

    if search_word in current_word.lower():
        found = True
        result_word = current_word
        print(f"üéØ –ù–∞–π–¥–µ–Ω–æ! –°–ª–æ–≤–æ '{result_word}' —Å–æ–¥–µ—Ä–∂–∏—Ç '{search_word}'")
    else:
        print(f"  ‚Üí –ù–µ –ø–æ–¥—Ö–æ–¥–∏—Ç, –ø—Ä–æ–¥–æ–ª–∂–∞–µ–º –ø–æ–∏—Å–∫...")
        index += 1  # –û–ë–Ø–ó–ê–¢–ï–õ–¨–ù–û –∏–∑–º–µ–Ω—è–µ–º —É—Å–ª–æ–≤–∏–µ!

if not found:
    print(f"‚ùå –°–ª–æ–≤–æ —Å –ø–æ–¥—Å—Ç—Ä–æ–∫–æ–π '{search_word}' –Ω–µ –Ω–∞–π–¥–µ–Ω–æ")

print(f"\n–†–µ–∑—É–ª—å—Ç–∞—Ç –ø–æ–∏—Å–∫–∞: {'–Ω–∞–π–¥–µ–Ω–æ' if found else '–Ω–µ –Ω–∞–π–¥–µ–Ω–æ'}")
print(f"–ü—Ä–æ—Å–º–æ—Ç—Ä–µ–Ω–æ —Å–ª–æ–≤: {index + 1 if found else len(words)}")
```

```python
# 7) —á–∏—Å—Ç–∫–∞ –ø—Ä–µ—Ñ–∏–∫—Å–∞
s = "...‚Äî¬´Amore¬ª"
while len(s) > 0 and not s[0].isalnum():
    s = s[1:]
s
```

```python
# 8) –Ω–∞–∫–∞–ø–ª–∏–≤–∞–µ–º –ø–µ—Ä–≤—ã–µ –±—É–∫–≤—ã
tmp = []
i = 0
while len(tmp) < 5 and i < len(words):
    if words[i]:
        tmp.append(words[i][0].lower())
    i += 1
tmp
```
