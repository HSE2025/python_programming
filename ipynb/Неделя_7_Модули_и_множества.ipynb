````python
from collections import Counter

stop_words = set(stopwords.words('russian'))

def analyze_text(text: str, title="–¢–µ–∫—Å—Ç"):
    """
    –ü–æ–ª–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Ç–µ–∫—Å—Ç–∞.

    Args:
        text (str): –¢–µ–∫—Å—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
        title (str): –ù–∞–∑–≤–∞–Ω–∏–µ —Ç–µ–∫—Å—Ç–∞

    Returns:
        dict: –°–ª–æ–≤–∞—Ä—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ –∞–Ω–∞–ª–∏–∑–∞
    """
    # –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞
    text = clean_text(text)
    words = text.lower().split()
    words = [word for word in words if word not in stop_words]

    # –ü–æ–¥—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç
    freq = Counter(words)

    # –°–æ–∑–¥–∞—ë–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –∫–∞–∫ —Å–ª–æ–≤–∞—Ä—å
    result = {
        "title": title,
        "tokens": len(words),           # –≤—Å–µ–≥–æ —Å–ª–æ–≤
        "types": len(set(words)),       # —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ª–æ–≤
        "ttr": len(set(words)) / len(words) if words else 0,  # Type-Token Ratio
        "vocab": set(words),            # –º–Ω–æ–∂–µ—Å—Ç–≤–æ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
        "freq": freq,                   # —á–∞—Å—Ç–æ—Ç—ã –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
        "top_10": freq.most_common(30), # —Ç–æ–ø-30 —Å–ª–æ–≤
        "hapax": [word for word, count in freq.items() if count == 1]  # —Å–ª–æ–≤–∞ —Å —á–∞—Å—Ç–æ—Ç–æ–π = 1
    }

    return result

print("‚úÖ –§—É–Ω–∫—Ü–∏—è analyze_text() —Å–æ–∑–¥–∞–Ω–∞!")
````

**–ß—Ç–æ –æ–±—ä–µ–¥–∏–Ω–∏–ª–∏:**
- **–°–ª–æ–≤–∞—Ä—å** (dict) ‚Äî —Å—Ç—Ä—É–∫—Ç—É—Ä–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞
- **Counter** ‚Äî —á–∞—Å—Ç–æ—Ç—ã —Å–ª–æ–≤
- **–ú–Ω–æ–∂–µ—Å—Ç–≤–æ** (set) ‚Äî —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ —Å–ª–æ–≤–∞
- **List comprehension** ‚Äî hapax legomena
- **–§—É–Ω–∫—Ü–∏—è** ‚Äî –ø–µ—Ä–µ–∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∫–æ–¥–∞

## 16. –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤

````python
def compare_texts(analysis1, analysis2):
    """
    –°—Ä–∞–≤–Ω–∏–≤–∞–µ—Ç –¥–≤–∞ –ø—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã—Ö —Ç–µ–∫—Å—Ç–∞.

    Args:
        analysis1 (dict): –†–µ–∑—É–ª—å—Ç–∞—Ç analyze_text() –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ —Ç–µ–∫—Å—Ç–∞
        analysis2 (dict): –†–µ–∑—É–ª—å—Ç–∞—Ç analyze_text() –¥–ª—è –≤—Ç–æ—Ä–æ–≥–æ —Ç–µ–∫—Å—Ç–∞

    Returns:
        dict: –°–ª–æ–≤–∞—Ä—å —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
    """
    vocab1 = analysis1["vocab"]
    vocab2 = analysis2["vocab"]

    comparison = {
        "common_words": vocab1 & vocab2,          # –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ
        "only_first": vocab1 - vocab2,            # —Ä–∞–∑–Ω–æ—Å—Ç—å
        "only_second": vocab2 - vocab1,           # —Ä–∞–∑–Ω–æ—Å—Ç—å
        "all_words": vocab1 | vocab2,             # –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ
        "jaccard": len(vocab1 & vocab2) / len(vocab1 | vocab2) if (vocab1 | vocab2) else 0
    }

    return comparison

print("‚úÖ –§—É–Ω–∫—Ü–∏—è compare_texts() —Å–æ–∑–¥–∞–Ω–∞!")
````

**–ß—Ç–æ –æ–±—ä–µ–¥–∏–Ω–∏–ª–∏:**
- **–û–ø–µ—Ä–∞—Ü–∏–∏ –º–Ω–æ–∂–µ—Å—Ç–≤** ‚Äî –≤—Å–µ 4 –æ–ø–µ—Ä–∞—Ü–∏–∏
- **–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ñ–∞–∫–∫–∞—Ä–∞** ‚Äî –º–µ—Ä–∞ –ø–æ—Ö–æ–∂–µ—Å—Ç–∏
- **–°–ª–æ–≤–∞—Ä—å** ‚Äî —Å—Ç—Ä—É–∫—Ç—É—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç

## 17. –§—É–Ω–∫—Ü–∏–∏ –¥–ª—è –≤—ã–≤–æ–¥–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

````python
def print_analysis(analysis):
    """
    –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤—ã–≤–æ–¥ –∞–Ω–∞–ª–∏–∑–∞ —Ç–µ–∫—Å—Ç–∞.
    """
    print(f"üìñ {analysis['title']}")
    print("=" * 60)
    print(f"–í—Å–µ–≥–æ —Å–ª–æ–≤ (tokens):     {analysis['tokens']}")
    print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã—Ö (types):      {analysis['types']}")
    print(f"TTR:                     {analysis['ttr']:.2f}")
    print(f"\n–¢–æ–ø-10 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤:")
    for word, count in analysis['top_10']:
        print(f"  {word:20} {count:3} —Ä–∞–∑(–∞)")
    print(f"\nHapax legomena:          {len(analysis['hapax'])} —Å–ª–æ–≤")
    if len(analysis['hapax']) > 0:
        print(f"–ü—Ä–∏–º–µ—Ä—ã: {', '.join(analysis['hapax'][:5])}")
    print()

def print_comparison(comparison, title1, title2):
    """
    –§–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤—ã–≤–æ–¥ —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–æ–≤.
    """
    print(f"üîç –°–†–ê–í–ù–ï–ù–ò–ï: '{title1}' vs '{title2}'")
    print("=" * 60)
    print(f"–û–±—â–∏—Ö —Å–ª–æ–≤:              {len(comparison['common_words'])}")
    print(f"–¢–æ–ª—å–∫–æ –≤ '{title1}':     {len(comparison['only_first'])}")
    print(f"–¢–æ–ª—å–∫–æ –≤ '{title2}':     {len(comparison['only_second'])}")
    print(f"–í—Å–µ–≥–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö:        {len(comparison['all_words'])}")
    print(f"\nüìä –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ñ–∞–∫–∫–∞—Ä–∞: {comparison['jaccard']:.2f}")

    # –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è
    jaccard = comparison['jaccard']
    if jaccard > 0.7:
        interpretation = "–û—á–µ–Ω—å –ø–æ—Ö–æ–∂–∏–µ —Ç–µ–∫—Å—Ç—ã"
    elif jaccard > 0.4:
        interpretation = "–ó–∞–º–µ—Ç–Ω–æ–µ —Å—Ö–æ–¥—Å—Ç–≤–æ"
    elif jaccard > 0.2:
        interpretation = "–£–º–µ—Ä–µ–Ω–Ω–æ–µ —Å—Ö–æ–¥—Å—Ç–≤–æ"
    else:
        interpretation = "–ù–∏–∑–∫–æ–µ —Å—Ö–æ–¥—Å—Ç–≤–æ"
    print(f"–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è: {interpretation}")
    print()

print("‚úÖ –§—É–Ω–∫—Ü–∏–∏ –≤—ã–≤–æ–¥–∞ —Å–æ–∑–¥–∞–Ω—ã!")
````

## 18. –ü—Ä–∏–º–µ–Ω—è–µ–º –≤—Å—é —Å–∏—Å—Ç–µ–º—É

````python
# –®–∞–≥ 1: –ü–æ–ª–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –∫–∞–∂–¥–æ–≥–æ —Ç–µ–∫—Å—Ç–∞
analysis1 = analyze_text(text1, "Harry Potter and the Philosopher's Stone")
analysis2 = analyze_text(text2, "Harry Potter and The Chamber of Secrets")

# –®–∞–≥ 2: –í—ã–≤–æ–¥ –∏–Ω–¥–∏–≤–∏–¥—É–∞–ª—å–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print_analysis(analysis1)
print_analysis(analysis2)
````
