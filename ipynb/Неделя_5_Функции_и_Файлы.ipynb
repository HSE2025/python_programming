## –§—É–Ω–∫—Ü–∏—è –º–æ–∂–µ—Ç –¥–µ–ª–∞—Ç—å –æ–±–∞ –¥–µ–π—Å—Ç–≤–∏—è

```python
def count_and_report(text):
    """–ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å–ª–æ–≤–∞ –∏ —Å–æ–æ–±—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç"""
    word_count = len(text.split())
    print(f"–ù–∞–π–¥–µ–Ω–æ —Å–ª–æ–≤: {word_count}")  # –ò–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –¥–ª—è –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    return word_count                      # –†–µ–∑—É–ª—å—Ç–∞—Ç –¥–ª—è –ø—Ä–æ–≥—Ä–∞–º–º—ã

count = count_and_report("–ú–∞–º–∞ –º—ã–ª–∞ —Ä–∞–º—É")
print(f"–°–æ—Ö—Ä–∞–Ω–µ–Ω–æ –≤ –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é: {count}")
```

## –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é

```python
def normalize(text, preserve_case=False, remove_punct=True):
    """–ù–æ—Ä–º–∞–ª–∏–∑—É–µ—Ç —Ç–µ–∫—Å—Ç —Å –Ω–∞—Å—Ç—Ä–∞–∏–≤–∞–µ–º—ã–º–∏ –æ–ø—Ü–∏—è–º–∏"""
    result = text.strip()

    if not preserve_case:
        result = result.lower()

    if remove_punct:
        result = result.replace(".", "").replace(",", "")

    return result

# –†–∞–∑–Ω—ã–µ —Å–ø–æ—Å–æ–±—ã –≤—ã–∑–æ–≤–∞
print(normalize("–¢–µ–∫—Å—Ç."))                          # —Ç–µ–∫—Å—Ç
```

## Type hints ‚Äî –ø–æ–¥—Å–∫–∞–∑–∫–∏ —Ç–∏–ø–æ–≤

```python
def count_vowels(text: str) -> int:
    """
    –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞—Å–Ω—ã—Ö –≤ —Ç–µ–∫—Å—Ç–µ.

    Args:
        text (str): –¢–µ–∫—Å—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞

    Returns:
        int: –ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞—Å–Ω—ã—Ö –±—É–∫–≤
    """
    vowels = "–∞–µ—ë–∏–æ—É—ã—ç—é—è–ê–ï–Å–ò–û–£–´–≠–Æ–Ø"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

result = count_vowels("–í–æ –≥–ª—É–±–∏–Ω–µ —Å–∏–±–∏—Ä—Å–∫–∏—Ö —Ä—É–¥")
print(f"–ì–ª–∞—Å–Ω—ã—Ö: {result}")
```

```python
lines_list = ["–í–æ –≥–ª—É–±–∏–Ω–µ —Å–∏–±–∏—Ä—Å–∫–∏—Ö —Ä—É–¥", "–•—Ä–∞–Ω–∏—Ç–µ –≥–æ—Ä–¥–æ–µ —Ç–µ—Ä–ø–µ–Ω—å–µ", "–ù–µ –ø—Ä–æ–ø–∞–¥–µ—Ç –≤–∞—à —Å–∫–æ—Ä–±–Ω—ã–π —Ç—Ä—É–¥", "–ò –¥—É–º –≤—ã—Å–æ–∫–æ–µ —Å—Ç—Ä–µ–º–ª–µ–Ω—å–µ"]

for line in lines_list:
    print(f"–ì–ª–∞—Å–Ω—ã—Ö: {count_vowels(line)}")
```

## –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏–π –ø—Ä–∏–º–µ—Ä: –Ω–∞–±–æ—Ä —Ñ—É–Ω–∫—Ü–∏–π –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ —Ç–µ–∫—Å—Ç–∞

```python
def count_vowels(text: str) -> int:
    """–ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç –≥–ª–∞—Å–Ω—ã–µ –ª–∞—Ç–∏–Ω—Å–∫–æ–≥–æ –∞–ª—Ñ–∞–≤–∏—Ç–∞"""
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
      if char in vowels:
        count += 1
    return count

def clean_text(text: str) -> str:
    """–û—á–∏—â–∞–µ—Ç —Ç–µ–∫—Å—Ç –æ—Ç –∑–Ω–∞–∫–æ–≤ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è"""
    for punct in ".,!?;:‚Äî-":
        text = text.replace(punct, "")
    return text.strip()

def get_word_count(text: str) -> int:
    """–ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ—Ç —Å–ª–æ–≤–∞ –ø–æ—Å–ª–µ –æ—á–∏—Å—Ç–∫–∏"""
    clean = clean_text(text)
    return len(clean.split())

def find_longest_word(text: str) -> str:
    """–ù–∞—Ö–æ–¥–∏—Ç —Å–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ"""
    clean = clean_text(text)
    words = clean.split()

    if not words:
        return ""

    return max(words, key=len)

# –¢–µ—Å—Ç–∏—Ä—É–µ–º
sample = "Nel mezzo del cammin di nostra vita"
print(f"Sample: {sample}")
print(f"–ì–ª–∞—Å–Ω—ã—Ö: {count_vowels(sample)}")
print(f"–°–ª–æ–≤: {get_word_count(sample)}")
print(f"–°–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ: {find_longest_word(sample)}")
```

## –û–±—ä–µ–¥–∏–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏–∏ –∏ —Ñ–∞–π–ª—ã ‚Äî –º–æ—â–Ω–∞—è –∫–æ–º–±–∏–Ω–∞—Ü–∏—è!

```python
def analyze_file(filename: str):
  """
  –ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç —Ç–µ–∫—Å—Ç–æ–≤—ã–π —Ñ–∞–π–ª –∏ –≤—ã–≤–æ–¥–∏—Ç –º–µ—Ç—Ä–∏–∫–∏.

  Args:
  filename: –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É
  """
  with open(filename, 'r', encoding='utf-8', errors='replace') as f:
    text = f.read()

  lines = text.splitlines()

  # –í—ã—á–∏—Å–ª—è–µ–º –º–µ—Ç—Ä–∏–∫–∏
  chars = len(text)
  lines_count = len(lines)
  words = len(text.split())
  avg_line_length = len(text.replace('\n', '')) / len(lines) if lines else 0
  vowels = count_vowels(text)
  longest_word = find_longest_word(text)

  # –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
  print("üìä –ê–Ω–∞–ª–∏–∑ —Ñ–∞–π–ª–∞:")
  print(f"–°–∏–º–≤–æ–ª–æ–≤: {chars}")
  print(f"–°—Ç—Ä–æ–∫: {lines_count}")
  print(f"–°–ª–æ–≤: {words}")
  print(f"–°—Ä–µ–¥–Ω—è—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏: {avg_line_length:.2f}")
  print(f"–ì–ª–∞—Å–Ω—ã—Ö: {vowels}")
  print(f"–°–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ: {longest_word}")

# –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ñ—É–Ω–∫—Ü–∏—é
analyze_file(filename)
```

```python
def poem_analysis_from_file(filename: str, title: str = None):
    """
    –ü–æ–ª–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Å—Ç–∏—Ö–æ—Ç–≤–æ—Ä–µ–Ω–∏—è –∏–∑ —Ñ–∞–π–ª–∞.

    Args:
        filename: –ü—É—Ç—å –∫ —Ñ–∞–π–ª—É —Å–æ —Å—Ç–∏—Ö–æ—Ç–≤–æ—Ä–µ–Ω–∏–µ–º
        title: –ù–∞–∑–≤–∞–Ω–∏–µ (–µ—Å–ª–∏ None, –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∏–º—è —Ñ–∞–π–ª–∞)
    """
    # –ß–∏—Ç–∞–µ–º —Ñ–∞–π–ª
    with open(filename, 'r', encoding='utf-8', errors='replace') as f:
        poem = f.read()

    # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∏–º—è —Ñ–∞–π–ª–∞ –∫–∞–∫ –∑–∞–≥–æ–ª–æ–≤–æ–∫, –µ—Å–ª–∏ –Ω–µ —É–∫–∞–∑–∞–Ω
    if title is None:
        title = filename

    # –ê–Ω–∞–ª–∏–∑ —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º –Ω–∞—à–∏—Ö —Ñ—É–Ω–∫—Ü–∏–π
    lines = poem.splitlines()
    clean = clean_text(poem)
    words = clean.split()

    # –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
    print(f"üìñ –ê–Ω–∞–ª–∏–∑ —Ñ–∞–π–ª–∞: {title}")
    print(f"="*50)
    print(f"üìè –°—Ç—Ä–æ–∫: {len(lines)}")
    print(f"üìù –°–ª–æ–≤: {len(words)}")
    print(f"üî§ –ì–ª–∞—Å–Ω—ã—Ö: {count_vowels(poem)}")
    print(f"üìè –°–∞–º–æ–µ –¥–ª–∏–Ω–Ω–æ–µ —Å–ª–æ–≤–æ: {find_longest_word(poem)}")
    print(f"üìè –°—Ä–µ–¥–Ω—è—è –¥–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏: {len(clean) / len(lines) if lines else 0:.2f}")
    print("="*50)
    print("‚úì –ê–Ω–∞–ª–∏–∑ –∑–∞–≤–µ—Ä—à—ë–Ω")

# –í—ã–∑—ã–≤–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é
poem_analysis_from_file(filename)
```
