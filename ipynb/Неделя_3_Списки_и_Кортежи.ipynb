```python
import requests

# 1) –≤—ã–±–µ—Ä–µ–º –∫–æ—Ä–ø—É—Å –∏ –≤–æ–∑—å–º—ë–º –ø–µ—Ä–≤—É—é –ø—å–µ—Å—É –∏–∑ —Å–ø–∏—Å–∫–∞
corpus = "rus"  # –º–æ–∂–Ω–æ –ø–æ–º–µ–Ω—è—Ç—å (eng, rus, ger, ita, fra, ...)
play_from_corpus = 1  # –º–æ–∂–Ω–æ –ø–æ–º–µ–Ω—è—Ç—å –Ω–∞ –¥—Ä—É–≥–æ–π –ø–æ—Ä—è–¥–∫–æ–≤—ã–π –Ω–æ–º–µ—Ä (–≤ —Ä–∞–º–∫–∞—Ö –¥–æ—Å—Ç—É–ø–Ω–æ–≥–æ –≤ –∫–æ—Ä–ø—É—Å–µ)

corpus_data = requests.get(f"https://dracor.org/api/v1/corpora/{corpus}").json()

if isinstance(corpus_data.get("plays"), list) and corpus_data["plays"]:
  choosen_play = corpus_data["plays"][play_from_corpus]
  play_title = choosen_play.get("title")
  play_id = choosen_play.get("id")
  print(f"–ü–µ—Ä–≤–∞—è –ø—å–µ—Å–∞ –≤ –∫–æ—Ä–ø—É—Å–µ {corpus_data.get('title')}: {play_title}")
else:
    play_id = None

```

```python
# 2) –∑–∞–≥—Ä—É–∑–∏–º –ø–µ—Ä—Å–æ–Ω–∞–∂–µ–π
characters = None
if play_id:
    url_chars = f"https://dracor.org/api/v1/corpora/{corpus}/plays/{play_id}/characters"
    characters = requests.get(url_chars).json()
characters[:2]
```

```python
# 3) —Å–æ–±–µ—Ä—ë–º —Å–ø–∏—Å–æ–∫ –∏–º—ë–Ω
names = [c.get("name", "") for c in characters if isinstance(c, dict)]
names = [n for n in names if n]  # —É–±–µ—Ä—ë–º –ø—É—Å—Ç—ã–µ
print(play_title, len(names))
print("–ü–µ—Ä–≤—ã–µ –∏–º–µ–Ω–∞:", sorted(names)[:10])
```
