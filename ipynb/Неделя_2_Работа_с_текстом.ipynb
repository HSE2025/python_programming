- **–°—Ç—Ä–æ–∫–∞ (`str`)** ‚Äî –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Å–∏–º–≤–æ–ª–æ–≤ Unicode. –ö–∞–≤—ã—á–∫–∏: `'...'`, `"..."`, –º–Ω–æ–≥–æ—Å—Ç—Ä–æ—á–Ω–æ `'''...'''` / `"""..."""`. –≠–∫—Ä–∞–Ω–∏—Ä–æ–≤–∞–Ω–∏–µ: `\n`, `\t`; ¬´—Å—ã—Ä—ã–µ¬ª —Å—Ç—Ä–æ–∫–∏: `r'...'`.
- **–°–∏–º–≤–æ–ª** ‚Äî —ç–ª–µ–º–µ–Ω—Ç —Å—Ç—Ä–æ–∫–∏. `len(s)` —Å—á–∏—Ç–∞–µ—Ç **—Å–∏–º–≤–æ–ª—ã**, –∞ –Ω–µ –±–∞–π—Ç—ã. –ò–Ω–¥–µ–∫—Å–∞—Ü–∏—è —Å 0; –æ—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω—ã–µ –∏–Ω–¥–µ–∫—Å—ã ‚Äî —Å –∫–æ–Ω—Ü–∞.
- **–ò–Ω–¥–µ–∫—Å / —Å—Ä–µ–∑** ‚Äî `s[i]`, `s[a:b:c]`. –ü—Ä–∞–≤–∞—è –≥—Ä–∞–Ω–∏—Ü–∞ **–Ω–µ –≤–∫–ª—é—á–∞–µ—Ç—Å—è**; —à–∞–≥ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é `1`; —Ä–∞–∑–≤–æ—Ä–æ—Ç: `s[::-1]`. –°—Ä–µ–∑—ã –±–µ–∑–æ–ø–∞—Å–Ω–µ–µ –∏–Ω–¥–µ–∫—Å–æ–≤.
- **–ò–º–º—É—Ç–∞–±–µ–ª—å–Ω–æ—Å—Ç—å** ‚Äî —Å—Ç—Ä–æ–∫–∏ **–Ω–µ –º–µ–Ω—è—é—Ç—Å—è –Ω–∞ –º–µ—Å—Ç–µ**. –ú–µ—Ç–æ–¥ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç **–Ω–æ–≤—É—é** —Å—Ç—Ä–æ–∫—É ‚Üí —Å–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç: `s = s.strip()`.
- **–ú–µ—Ç–æ–¥ vs —Ñ—É–Ω–∫—Ü–∏—è** ‚Äî –º–µ—Ç–æ–¥ —É –æ–±—ä–µ–∫—Ç–∞: `s.lower()`, —Ñ—É–Ω–∫—Ü–∏—è ¬´—Å–Ω–∞—Ä—É–∂–∏¬ª: `len(s)`.

### –ö—Ä–∞—è, –ø—Ä–æ–±–µ–ª—ã, —Ä–∞–∑–±–∏–µ–Ω–∏–µ
- **–ö—Ä–∞–µ–≤—ã–µ –ø—Ä–æ–±–µ–ª—ã** ‚Äî `s.strip()`, `s.lstrip()`, `s.rstrip()`; –Ω–∞–±–æ—Ä —Å–∏–º–≤–æ–ª–æ–≤: `s.strip(' .,!')`.
- **–ù–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—è –≤–Ω—É—Ç—Ä–µ–Ω–Ω–∏—Ö –ø—Ä–æ–±–µ–ª–æ–≤** ‚Äî `' '.join(s.split())`.
- **–†–∞–∑–±–∏–µ–Ω–∏–µ/—Å–±–æ—Ä–∫–∞** ‚Äî `s.split()` (—Å—Ö–ª–æ–ø—ã–≤–∞–µ—Ç —Å–µ—Ä–∏–∏ –ø—Ä–æ–±–µ–ª–æ–≤), `s.split(' ')` (—Å—Ç—Ä–æ–≥–æ –ø–æ –æ–¥–Ω–æ–º—É –ø—Ä–æ–±–µ–ª—É), `s.splitlines()`.  
  –°–±–æ—Ä–∫–∞: `' '.join(words)` ‚Äî **—ç–ª–µ–º–µ–Ω—Ç—ã –¥–æ–ª–∂–Ω—ã –±—ã—Ç—å —Å—Ç—Ä–æ–∫–∞–º–∏**.

### –ó–∞–º–µ–Ω—ã, –ø–æ–∏—Å–∫, –ø–æ–∑–∏—Ü–∏–∏
- **–ó–∞–º–µ–Ω—ã** ‚Äî `s.replace(old, new[, count])` ‚Äî –ø–æ–¥—Å—Ç—Ä–æ—á–Ω–∞—è –∑–∞–º–µ–Ω–∞, **—É—á–∏—Ç—ã–≤–∞–µ—Ç —Ä–µ–≥–∏—Å—Ç—Ä**, –Ω–µ –∑–Ω–∞–µ—Ç –ø—Ä–æ –≥—Ä–∞–Ω–∏—Ü—ã —Å–ª–æ–≤–∞.
- **–ü–æ–∏—Å–∫/—Å—á—ë—Ç—á–∏–∫–∏** ‚Äî `sub in s`, `s.count(sub)`.
- **–ù–∞—á–∞–ª–æ/–∫–æ–Ω–µ—Ü** ‚Äî `s.startswith(pref)`, `s.endswith(suf)`.
- **–ü–æ–∑–∏—Ü–∏–∏** ‚Äî `s.find(sub)` / `s.rfind(sub)` (‚àí1 –µ—Å–ª–∏ –Ω–µ—Ç), `s.index(sub)` ‚Äî –∫–∏–¥–∞–µ—Ç –æ—à–∏–±–∫—É.

### –ü—Ä–æ–≤–µ—Ä–∫–∞ —Å–æ–¥–µ—Ä–∂–∏–º–æ–≥–æ –∏ —Ä–µ–≥–∏—Å—Ç—Ä
- **–ü—Ä–æ–≤–µ—Ä–∫–∏** ‚Äî `s.isdigit()`, `s.isalpha()`, `s.isalnum()`, `s.isspace()`, `s.isupper()`, `s.islower()`.
- **–†–µ–≥–∏—Å—Ç—Ä** ‚Äî `s.lower()`, `s.upper()`, `s.capitalize()`, `s.title()` (‚ö† –º–æ–∂–µ—Ç –ø–æ—Ä—Ç–∏—Ç—å —Å–ª–æ–≤–∞ —Å –∞–ø–æ—Å—Ç—Ä–æ—Ñ–∞–º–∏).
- **–†–µ–≥–∏—Å—Ç—Ä–æ–Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–π –ø–æ–∏—Å–∫** ‚Äî `needle.lower() in hay.lower()`.
