### –°–ª–æ–≤–∞—Ä–∏ (dict)
- **–°–æ–∑–¥–∞–Ω–∏–µ**: `{}`, `dict()`, `{"key": "value"}`
- **–î–æ—Å—Ç—É–ø**: `d["key"]` (–ø—Ä—è–º–æ–π), `d.get("key", default)` (–±–µ–∑–æ–ø–∞—Å–Ω—ã–π)

## 9. –†—É—á–Ω–æ–π –ø–æ–¥—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç (–ø–∞—Ç—Ç–µ—Ä–Ω –Ω–∞–∫–æ–ø–ª–µ–Ω–∏—è)

```python
# –¢–µ–∫—Å—Ç –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
text = """
Not for the first time, an argument had broken out over breakfast at number four, Privet Drive. Mr. Vernon Dursley had been woken in the early hours of the morning by a loud, hooting noise from his nephew Harry‚Äôs room.
"""

# –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞
clean_text = text.lower()
for punct in ",.!?;:‚Äî-":
    clean_text = clean_text.replace(punct, "")

words = clean_text.split()
print(f"–í—Å–µ–≥–æ —Å–ª–æ–≤: {len(words)}")
print(f"–ü–µ—Ä–≤—ã–µ 10 —Å–ª–æ–≤: {words[:10]}")
```

```python
# –†—É—á–Ω–æ–π –ø–æ–¥—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç ‚Äî –ü–ê–¢–¢–ï–†–ù –ù–ê–ö–û–ü–õ–ï–ù–ò–Ø
freq = {}

for word in words:
    # –ï—Å–ª–∏ —Å–ª–æ–≤–æ —É–∂–µ –µ—Å—Ç—å ‚Äî —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º —Å—á—ë—Ç—á–∏–∫
    # –ï—Å–ª–∏ –Ω–µ—Ç ‚Äî —Å–æ–∑–¥–∞—ë–º —Å –Ω–∞—á–∞–ª—å–Ω—ã–º –∑–Ω–∞—á–µ–Ω–∏–µ–º 1
    # test = freq.get(word, 0)
    # print(f"Test new: ", word)
    # print(f"Meet:", test)
    freq[word] = freq.get(word, 0) + 1

print("–ß–∞—Å—Ç–æ—Ç–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å:")
for word, count in freq.items():
    print(f"  {word}: {count}")
```

```python
# –†–∞–∑–±–æ—Ä –ø–∞—Ç—Ç–µ—Ä–Ω–∞ (–º–µ–¥–ª–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è –¥–ª—è –ø–æ–Ω–∏–º–∞–Ω–∏—è)
freq_slow = {}

for word in words:
    if word in freq_slow:
        # –°–ª–æ–≤–æ —É–∂–µ –≤—Å—Ç—Ä–µ—á–∞–ª–æ—Å—å ‚Äî —É–≤–µ–ª–∏—á–∏–≤–∞–µ–º
        freq_slow[word] = freq_slow[word] + 1
    else:
        # –ü–µ—Ä–≤–æ–µ –≤—Ö–æ–∂–¥–µ–Ω–∏–µ ‚Äî —Å–æ–∑–¥–∞—ë–º —Å 1
        freq_slow[word] = 1

print("\n–¢–æ—Ç –∂–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç (–º–µ–¥–ª–µ–Ω–Ω–∞—è –≤–µ—Ä—Å–∏—è):")
print(freq_slow)
```

```python
# –ø–æ–≤—Ç–æ—Ä–∏—Ç–µ —á–∞—Å—Ç–æ—Ç–Ω—ã–π –∞–Ω–∞–ª–∏–∑ —Å –≤–∞—à–∏–º —Ç–µ–∫—Å—Ç–æ–º
```

## 10. –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ —á–∞—Å—Ç–æ—Ç–µ

```python
# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ —á–∞—Å—Ç–æ—Ç–µ (–æ—Ç —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö –∫ —Ä–µ–¥–∫–∏–º)
sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)

print("–¢–æ–ø-5 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤:")
for word, count in sorted_freq[:5]:
    print(f"  {word}: {count}")
```

```python
# –°–æ—Ä—Ç–∏—Ä–æ–≤–∫–∞ –ø–æ –∞–ª—Ñ–∞–≤–∏—Ç—É
sorted_alpha = sorted(freq.items())

print("\n–ü–æ –∞–ª—Ñ–∞–≤–∏—Ç—É:")
for word, count in sorted_alpha[:5]:
    print(f"  {word}: {count}")
```

## 11. Collections.Counter ‚Äî –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –ø–æ–¥—Å—á—ë—Ç

```python
from collections import Counter

# –ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –ø–æ–¥—Å—á—ë—Ç —á–∞—Å—Ç–æ—Ç
freq_counter = Counter(words)

print("–ß–∞—Å—Ç–æ—Ç–Ω—ã–π —Å–ª–æ–≤–∞—Ä—å (Counter):")
print(freq_counter)
```

```python
# –¢–æ–ø-5 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤
print("–¢–æ–ø-5 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö —Å–ª–æ–≤:")
for word, count in freq_counter.most_common(5):
    print(f"  {word}: {count}")
```

## 14. Counter: –æ–ø–µ—Ä–∞—Ü–∏–∏ —Å–æ —Å–ª–æ–≤–∞—Ä—è–º–∏ —á–∞—Å—Ç–æ—Ç

```python
text1 = "—Ä–æ–∑–∞ —Ä–æ–∑–∞ —Ä–æ–∑–∞ –ª–∏–ª–∏—è –ª–∏–ª–∏—è —Ç—é–ª—å–ø–∞–Ω"
text2 = "—Ä–æ–∑–∞ —Ä–æ–∑–∞ –ª–∞–Ω–¥—ã—à –ª–∞–Ω–¥—ã—à –ª–∞–Ω–¥—ã—à"

counter1 = Counter(text1.split())
counter2 = Counter(text2.split())

print("–¢–µ–∫—Å—Ç 1:", counter1)
print("–¢–µ–∫—Å—Ç 2:", counter2)
```

```python
# –°–ª–æ–∂–µ–Ω–∏–µ ‚Äî –æ–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ –∫–æ—Ä–ø—É—Å–æ–≤
combined = counter1 + counter2
print("–°–ª–æ–∂–µ–Ω–∏–µ:", combined)
```

```python
# –í—ã—á–∏—Ç–∞–Ω–∏–µ ‚Äî —É–Ω–∏–∫–∞–ª—å–Ω—ã–µ –¥–ª—è –ø–µ—Ä–≤–æ–≥–æ
diff = counter1 - counter2
print("–í—ã—á–∏—Ç–∞–Ω–∏–µ (1 - 2):", diff)
```

```python
# –ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (–º–∏–Ω–∏–º—É–º) ‚Äî –æ–±—â–∏–µ —Å–ª–æ–≤–∞
intersection = counter1 & counter2
print("–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ (min):", intersection)
```

```python
# –û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (–º–∞–∫—Å–∏–º—É–º)
union = counter1 | counter2
print("–û–±—ä–µ–¥–∏–Ω–µ–Ω–∏–µ (max):", union)
```

```python
dict1 = {"title": "–ë–µ–¥–Ω–∞—è –õ–∏–∑–∞", "author": "–ö–∞—Ä–∞–º–∑–∏–Ω"}
dict2 = {"year": 1792, "genre": "–ø–æ–≤–µ—Å—Ç—å"}

# –°–ø–æ—Å–æ–± 1: .update() (–∏–∑–º–µ–Ω—è–µ—Ç –ø–µ—Ä–≤—ã–π —Å–ª–æ–≤–∞—Ä—å)
dict1_copy = dict1.copy()
dict1_copy.update(dict2)
print("update():", dict1_copy)
```
